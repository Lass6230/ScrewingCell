```python
import matplotlib.pyplot as plt
import csv

# Reading in data from CSV file
time = []
torque = []
with open(r'C:\Users\GHB\Desktop\Final test\Intrinsic data\N\i250420231005.csv', 'r') as csv_file:
    csv_reader = csv.reader(csv_file)
    next(csv_reader)  # Skip header row
    for row in csv_reader:
        time.append(float(row[0]))
        torque.append(float(row[2]))

# Plotting the data
fig, ax = plt.subplots(figsize=(8, 5), dpi=120)
ax.plot(time, torque, linewidth=0.5, color='blue', marker='o', markersize=0.01)

# Adding labels and title
ax.set_xlabel('Depth (mm)', fontsize=12)
ax.set_ylabel('Torque (Nm)', fontsize=12)
#ax.set_title('Torque vs De for Screwdriving Process', fontsize=14)

# Setting x and y axis limits
ax.set_xlim([min(time), max(time)])
ax.set_ylim([0, max(torque)+1])

# Adding grid lines
ax.grid(True)

# Adding legend
ax.legend(['Torque'], loc='upper right', fontsize=12)

# Removing top and right spines
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

# Show the plot
plt.show()
```

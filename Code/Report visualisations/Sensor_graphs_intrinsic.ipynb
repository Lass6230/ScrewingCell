```python
# Torque graph by class

import pandas as pd
import matplotlib.pyplot as plt
import os

# list of files
files = [r"C:\Users\GHB\Desktop\SCREW PROJECT\Machine Learning\Final data\Intrinsic data\N\i030520235006.csv",
         r"C:\Users\GHB\Desktop\SCREW PROJECT\Machine Learning\Final data\Intrinsic data\NS\i030520235038.csv",
         r"C:\Users\GHB\Desktop\SCREW PROJECT\Machine Learning\Final data\Intrinsic data\OT\i050520238000.csv",
         r"C:\Users\GHB\Desktop\SCREW PROJECT\Machine Learning\Final data\Intrinsic data\P\i030520234012.csv",
         r"C:\Users\GHB\Desktop\SCREW PROJECT\Machine Learning\Final data\Intrinsic data\UT\i030520234000.csv"]

plt.figure(figsize=(10,5))

# choose a color cycle for the plots, one color for each file
colors = ['b', 'g', 'r', 'c', 'm']

for file, color in zip(files, colors):
    df = pd.read_csv(file)
    # plot the torque as a function of time
    foldername = os.path.basename(os.path.dirname(file))  # get folder name
    plt.plot(df['Time (ms)']/1000, df['Torque (Nm)'], color=color, label=foldername)

plt.xlabel('Time (s)', fontsize=14)  # increased font size
plt.ylabel('Torque (Nm)', fontsize=14)  # increased font size
plt.title('Torque as a Function of Time', fontsize=16)  # increased font size

# increase x-tick and y-tick label sizes
plt.tick_params(axis='both', which='major', labelsize=12)

# add a box around the plot
#plt.box(on=True)

# add a grid
plt.grid(True)

# add a legend
plt.legend()
plt.show()
```
